```python
#import dependecies
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import sqlalchemy 
from sqlalchemy import create_engine
import scipy.stats as sts
```

```python
#install pscycopyg2
pip install psycopg2-binary
```

```python
#create engine link
engine = create_engine('postgresql://postgres:postgres@localhost:5432/SQL_HW_DB')
```

```python
# connect to engine
conn = engine.connect()
```

```python
#read salary table
salary_df = pd.read_sql("SELECT * FROM salaries",conn)
salary_df.head()
```

```python
# select only salaries 
salary = salary_df["salary"]
sts.mode(salary)
```

```python
#create histogram for salaries
plt.hist(salary)
plt.xlabel("Salaries ($)")
plt.ylabel("Counts")
plt.title("Salary Ranges for Employees")
plt.savefig("bonus_histogram.png", bbox_inches='tight',dpi=100)
plt.show()
```

```python
# read title table
title_df = pd.read_sql("SELECT * FROM titles",conn)
title_df
```

```python
# read employee table
employee_df = pd.read_sql("SELECT * FROM employees",conn)
employee_df.columns
```

```python
# select necessary columns
employee_df = employee_df[["emp_no","emp_title_id"]]
employee_df
```

```python
#rename so can merge
clean_employee = employee_df.rename(columns={"emp_title_id":"title_id"})
```

```python
# merge employee and title df
merge_1 = pd.merge(clean_employee,title_df,on="title_id",how="inner")
merge_1
```

```python
# merge previously merge df with salary df
final_df = pd.merge(merge_1,salary_df,on="emp_no",how="inner")
final_df
```

```python
# groupby title and get average 
groupby_df = final_df.groupby("title").mean()
```

```python
# remove unnecessary columns
groupby_df = groupby_df.drop(columns=["emp_no"])
groupby_df
```

```python
# round salaries 
groupby_df["salary"] = groupby_df["salary"].map("{:.2f}".format)
groupby_df
```

```python
# reset the index and convert datatype
clean_df = groupby_df.reset_index()
clean_df["salary"] = clean_df["salary"].astype(float)
clean_df
```

```python
# create bar chart 
x_axis = clean_df["title"]
y_axis = clean_df["salary"]
plt.bar(x_axis,y_axis, align="center",color=["b","r","orange","purple","y","green","pink"],alpha=0.6)
plt.xticks(x_axis,rotation=90)
plt.ylabel("Salary ($)")
plt.xlabel("Employee Title")
plt.title("Average Salary by Title")
plt.savefig("bonus_barchart.png", bbox_inches='tight',dpi=100)
plt.show()
```
